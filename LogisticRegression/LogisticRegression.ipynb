```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("treated_data/bank_train_treated.csv")
df
```

```python
#Transformando features categóricas em numéricas utilizando LabelEncoder
categoricas = df.select_dtypes("object").columns
print(categoricas)
```

```python
for i in range(len(categoricas)):
    le = LabelEncoder()
    df[categoricas[i]] = le.fit_transform(df[categoricas[i]])
    #print(df[categoricas[i]])
df
```

```python
X = df.copy().drop(columns=["y"])
Y = df["y"]
```

```python
Y
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=1234)
X_train_scaled = StandardScaler().fit_transform(X_train)
X_test_scaled = StandardScaler().fit_transform(X_test)
models = []
predictions = []
scores = []
max_score = 0
index_max = 0
for i in range(1000):
    models.append(LogisticRegression(max_iter=i))#Testing newton-cg solver
    models[i].fit(X_train_scaled, Y_train)
    predictions.append(models[i].predict(X_test_scaled))
    scores.append(models[i].score(X_test_scaled, Y_test))
for i in range(len(models)):
    if scores[i] > max_score:
        max_score = scores[i]
        index_max = i
    print(scores[i], i)
print("Max score:", max_score, "Max_iter: ", index_max)
```

```python
#Com o melhor modelo em mãos o aplicaremos no dataset de teste
best_model = models[index_max-100]
df_test = pd.read_csv("treated_data/bank_test.csv")
df_test
```

```python
df_test.isna().sum()
```

```python
ids = df_test["id"]
df_test.drop(columns=["id"], inplace=True)
```

```python
df_test
```

```python
df_test.drop(columns=["location"], inplace=True)
categoricas_test = df_test.select_dtypes("object").columns
numericas_test = df_test.select_dtypes("number").columns
print(categoricas_test, numericas_test)
```

```python

def print_unique(data=None):
    for col in data.columns:
        print("Col:", col,"\n", data[col].unique(), data[col].dtype)
print_unique(df_test)
```

```python
df_test.replace({"admin.": "administrator"}, inplace=True)
```

```python
for numeric in numericas_test:
    if df_test[numeric].isna().sum() > 0:
        if df_test[numeric].dtype == "int64" or numeric in ["age", "previous", "campaign"]:
            df_test[numeric] = df_test[numeric].fillna(np.int64(df_test[numeric].mean()))
        else:
            df_test[numeric] = df_test[numeric].fillna(df_test[numeric].mean())
for category in categoricas_test:
    if df_test[category].isna().sum() > 0:
        df_test[category] = df_test[category].fillna(df_test[category].mode()[0])
    
print_unique(df_test)
```

```python
df_test.replace({"n": "no", "y": "yes"}, inplace=True)
```

```python
print_unique(df_test)
```

```python
for i in range(len(categoricas_test)):
    le = LabelEncoder()
    df_test[categoricas_test[i]] = le.fit_transform(df_test[categoricas_test[i]])
print_unique(df_test)
```

```python
df_test_scaled = StandardScaler().fit_transform(df_test)
```

```python
df_test_scaled
```

```python
target_pred = best_model.predict(df_test_scaled)
```

```python
target = pd.DataFrame()
target["id"] = ids
target["y"] = target_pred
print_unique(target)
```

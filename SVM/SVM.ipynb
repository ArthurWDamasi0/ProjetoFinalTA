```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv("./Limpos/df_final.csv")
df_test = pd.read_csv("./Limpos/df_teste_final.csv")
df
```

###Treinando o modelo

```python
#separando o df
X = df.drop('y', axis = 1)
y = df['y']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 109)
```

```python
#Import svm model
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

# Model Precision: what percentage of positive tuples are labeled as such?
print("Precision:",metrics.precision_score(y_test, y_pred))

# Model Recall: what percentage of positive tuples are labelled as such?
print("Recall:",metrics.recall_score(y_test, y_pred))
```

```python
y_pred = clf.predict(df_test)
df_resultado = pd.read_csv("./raw_data/bank_test")
```

```python
df_resultado = pd.DataFrame({'id': np.array(df_resultado['id']), "y": y_pred})
df_resultado.index = df_resultado["id"]
df_resultado = df_resultado.drop('id',axis=1)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.preprocessing import StandardScaler 
from sklearn.svm import SVC 
from imblearn.over_sampling import SMOTE
```

```python
df = pd.read_csv("./data_treated.csv")
df_test = pd.read_csv("./data_treated_final.csv")
df
```

```python
df_test
```

```python
X = df.drop('y', axis = 1)
y = df['y']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

smote = SMOTE()
X_train, y_train = smote.fit_resample(X_train, y_train)

scaler = StandardScaler().fit(X_train) 
X_train_scaled = scaler.transform(X_train) 
X_test_scaled = scaler.transform(X_test) 
```

```python
for c in [0.125,0.25,2,4,100,1000,0.1,1,10,0.01,0.001,0.0001]:
    for gama in [0.125,0.25,2,4,100,1000,0.1,1,10,0.01,0.001,0.0001]:
        model = SVC(kernel='rbf', C=c, gamma=gama) 
        model.fit(X_train_scaled, y_train)

        y_pred = model.predict(X_test_scaled) 
        accuracy = metrics.accuracy_score(y_test, y_pred) 
        precision = metrics.precision_score(y_test, y_pred)
        recall = metrics.recall_score(y_test, y_pred)

        print(gama, model)
        print("Precision:",precision)
        print('Accuracy:', accuracy) 
        print("Recall:",recall) 
        print()
```

```python
new_data_scaled = scaler.transform(df_test) 
y_pred = model.predict(new_data_scaled)
```

```python
df_resultado = pd.DataFrame({'id': np.array(df_test['id']), "y": y_pred})
df_resultado.index = df_resultado["id"]
df_resultado = df_resultado.drop('id',axis=1)
df_resultado.to_csv("./target.csv")
```

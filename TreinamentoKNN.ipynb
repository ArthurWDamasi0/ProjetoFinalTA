```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv("C:\\Users\\chase\\Documents\\GitHub\\ProjetoFinalTA\\df_final.csv")
df_test = pd.read_csv("C:\\Users\\chase\\Documents\\GitHub\\ProjetoFinalTA\\df_teste_final.csv")
df
```

###Tratando o df

```python
#preparando o datasset para utilizar knn
# df = df.drop('poutcome', axis=1)
# df = df[df['poutcome'] != 'unknown']
# # df = df[df['poutcome'] != 'other']
# df = df[df['contact'] != 'unknown']
# df = df[df['job'] != 'unknown']
# df = df[df['education'] != 'unknown']
df = df.dropna()
```

```python
#transformando dados categóricos com get_dummies
df = pd.get_dummies(df)
df_test = pd.get_dummies(df_test)
df
```

###Colocando o df na escala correta

```python
#botando o dataframe na escala para o knn
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)
scaled_data
df = pd.DataFrame(scaled_data, columns = df.columns)
df.head()

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df_test)
scaled_data
df_test = pd.DataFrame(scaled_data, columns = df_test.columns)
df_test.head()
```

###Treinando o modelo

```python
#separando o df
X = df.drop('y', axis = 1)
y = df['y']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 5)

knn = KNeighborsClassifier(n_neighbors = 20)
knn.fit(X_train, y_train)
knn.score(X_test , y_test)
```

```python
y_pred = knn.predict(df_test)
df_resultado = pd.read_csv("C:\\Users\\chase\\Documents\\GitHub\\ProjetoFinalTA\\raw_data\\bank_test.csv")
```

```python
df_resultado = pd.DataFrame({'id': np.array(df_resultado['id']), "y": y_pred})
df_resultado.index = df_resultado["id"]
df_resultado = df_resultado.drop('id',axis=1)
```
